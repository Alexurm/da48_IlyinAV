# Определение перспективного тарифа для телеком-компании

Компания «Мегалайн» — федеральный оператор сотовой связи. Клиентам предлагаются два тарифных плана: «Смарт» и «Ультра».
Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.  
В рамках данной работы необходимо выполнить предварительный анализ тарифов на небольшой выборке клиентов (500 пользователей "Мегалайна") и сделать вывод - какой тариф является более предпочтительным.

В исследование входят следующие шаги:  
1. изучение данных и их предобработка;
2. расчеты и добавление результатов в таблицу;
2. анализ данных;
3. проверка гипотез.

Для каждого пользователя тарифов необходимо посчитать:
* количество сделанных звонков и израсходованных минут разговора по месяцам;
* количество отправленных сообщений по месяцам;
* объем израсходованного интернет-трафика по месяцам;
* помесячную выручку с каждого пользователя.

В рамках анализа данных необходимо описать поведение клиентов оператора, а именно ответить на вопросы: сколько минут разговора, сколько сообщений и какой объём интернет-трафика требуется пользователям каждого тарифа в месяц.  
Требуется рассчитать среднее количество, дисперсию и стандартное отклонение, построить гистограммы и описать полученные распределения.

Также требуется выполнить проверку следующих гипотез:
* средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
* средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

# Описание тарифов

**Тариф "Смарт"**

1. Ежемесячная плата: 550 рублей
2. Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика
3. Стоимость услуг сверх тарифного пакета:
* минута разговора: 3 рубля
* сообщение: 3 рубля
* 1 Гб интернет-трафика: 200 рублей

**Тариф "Ультра"**

1. Ежемесячная плата: 1950 рублей
2. Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика
3. Стоимость услуг сверх тарифного пакета:
* минута разговора: 1 рубль
* сообщение: 1 рубль
* 1 Гб интернет-трафика: 150 рублей

**Примечание**  
«Мегалайн» всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта. Принимается, что неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.

### Расчет помесячной выручки

Для дальнейшего анализа данных потребуется рассчитать помесячную выручку с каждого пользователя, а именно выполнить следующие шаги:
1. вычесть бесплатный лимит из суммарного количества звонков, сообщений и интернет трафика;
2. остаток умножить на значение из тарифного плана;
3. прибавить абонентскую плату, соответствующую тарифному плану.  

Вызовем таблицу с тарифами:

```python
tariffs
```

```python
# функция для расчета помесячной выручки:
def revenue_per_month(row):
    # определение переменных:
    calls_duration = row['calls_duration']
    messages_count = row['messages_count']
    gb_used = row['gb_used']
    tariff = row['tariff']
    # проверка условия тарифа "Смарт":
    if tariff == 'smart':
        # перерасход количества минут сверх включенного лимита:
        if calls_duration > 500:
            calls_duration_over_limit = calls_duration - 500
            calls_over_limit_price = calls_duration_over_limit * 3
        else:
            calls_over_limit_price = 0
        # перерасход количества сообщений сверх включенного лимита:
        if messages_count > 50:
            messages_count_over_limit = messages_count - 50
            messages_over_limit_price = messages_count_over_limit * 3
        else:
            messages_over_limit_price = 0
        # перерасход израсходованного интернет трафика:
        if gb_used > 15:
            gb_used_over_limit = gb_used - 15
            gb_over_limit_price = gb_used_over_limit * 200
        else:
            gb_over_limit_price = 0
        
        # ежемесячная плата по тарифу "Смарт":
        rub_monthly_fee = 550 
        
        # полная выручка для тарифа "Смарт":
        revenue_per_month = calls_over_limit_price + messages_over_limit_price + gb_over_limit_price + rub_monthly_fee
        
        return revenue_per_month
    
    elif tariff == 'ultra':
        # перерасход количества минут сверх включенного лимита:
        if calls_duration > 3000:
            calls_duration_over_limit = calls_duration - 3000
            calls_over_limit_price = calls_duration_over_limit * 1
        else:
            calls_over_limit_price = 0
        # перерасход количества сообщений сверх включенного лимита:
        if messages_count > 1000:
            messages_count_over_limit = messages_count - 1000
            messages_over_limit_price = messages_count_over_limit * 1
        else:
            messages_over_limit_price = 0
        # перерасход израсходованного интернет трафика:
        if gb_used > 30:
            gb_used_over_limit = gb_used - 30
            gb_over_limit_price = gb_used_over_limit * 150
        else:
            gb_over_limit_price = 0
        
        # ежемесячная плата по тарифу "Ультра":
        rub_monthly_fee = 1950 
        
        # полная выручка для тарифа "Ультра":
        revenue_per_month = calls_over_limit_price + messages_over_limit_price + gb_over_limit_price + rub_monthly_fee
        
        return revenue_per_month
    
```

```python
general_pivot['revenue_per_month'] = general_pivot.apply(revenue_per_month, axis=1)
```

* `messages_pivot` (количество отправленных сообщений по месяцам для каждого пользователя) 
* `interner_pivot` (объем израсходованного интернет-трафика по месяцам для каждого пользователя в гигабайтах)

Проверим правило трех сигм, чтобы понять - входят ли значения `internet_ultra` в промежуток по трем стандартным отклонениям в обе стороны:

```python
# счетчик значений для подсчета попаданий в диапазон трех стандартных отклонений в обе стороны:
count = 0

# нижняя граница диапазона:
internet_ultra_bottom_line = internet_ultra_mean - 3 * internet_ultra_standard_deviation
# верхняя граниза диапазона:
internet_ultra_top_line = internet_ultra_mean + 3 * internet_ultra_standard_deviation

# цикл для перебора значений:
for col_name, data in internet_ultra.items():
    if data > internet_ultra_bottom_line and data < internet_ultra_top_line:
        count += 1

# правило трех сигм - 99% всех значений находятся в промежутке шести стандартных отклонений:
if count / len(internet_ultra) >= 0.99:
    internet_ultra_3_sigma = 'Выполняется'
    print ('Правило трех сигм соблюдается, изучаемая величина распределена нормально.')
else:
    internet_ultra_3_sigma = 'Не выполняется'
    print ('Правило трех сигм не соблюдается, изучаемая величина не распределена нормально.')
```

Это означает, что есть наличие выбросов вне интервала. Рассмотрим значение выше границы нормального распределения:

```python
internet_ultra[internet_ultra > internet_ultra_top_line].sort_values()
```

Всего одно такое значение.

## Проверка гипотез

Необходимо проверить следующие гипотезы:
* средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются;
* средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

*Немного теории:*  

Проверка гипотез начинается с формулировки нулевой гипотезы H₀.  
H₀ всегда формулируется так, чтобы использовать знак равенства. Исходя из H₀ формулируется альтернативная гипотеза H₁. Это утверждение, которое принимается верным, если отбрасывается H₀. Иными словами, H₁ для H₀ — это альтернативная гипотеза, которая противоречит нулевой гипотезе.

При сравнении двух выборок проверяемая нулевая гипотеза состоит в том, что обе эти выборки происходят из нормального распределенных генеральных совокупностей с одинаковыми средними значениями:  
*H₀: μ₁=μ₁*

В общем виде проверка этой гипотезы выполняется при помощи p-value (уровень значимости статистики разности), который оценивает вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна. p-value - это вероятность, которая приводит к результатам, таким же предельным, какими могут оказаться наблюдаемые результаты с учетом модели нулевой гипотезы.

p-value сравнивается с выбранным уровнем значимости α (альфа), общепринятые пороговые значения - 5% и 1%, я буду принимать **5%**. Выбор любого из двух является произвольным решением - в данной процедуре нет ничего, что будет гарантировать правильные решения в x% случаях.
Значение α (уровень значимости) - это порог "необычности" в случайной модели нулевой гипотезы, по которому мы либо не будем отклонять нулевую гипотезу, либо наоборот отклоним и выберем альтернативную гипотезу.

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам:  
`scipy.stats.ttest_ind (array1, array2, equal_var)`  
Импортируем библиотеку:

```python
# импорт библиотеки scipy.stats:
from scipy import stats as st
```

### Проверка гипотезы об отличии средней выручки по тарифам «Ультра» и «Смарт»

Необходимо проверить, действительно ли средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются.

**Нулевая гипотеза H₀**  
средняя выручка пользователей тарифов «Ультра» и «Смарт» одинаковая.  
**Альтернативная гипотеза H₁**  
средняя выручка пользователей тарифов «Ультра» и «Смарт» различается.

Сгруппируем пользователей по тарифам, создав два массива `revenue_per_month_smart` (выручка по тарифу «Смарт») и `revenue_per_month_ultra` (выручка по тарифу «Ультра»):

```python
revenue_per_month_smart = general_pivot[general_pivot['tariff'] == 'smart']['revenue_per_month']
```

```python
revenue_per_month_ultra = general_pivot[general_pivot['tariff'] == 'ultra']['revenue_per_month']
```

Рассмотрим средние значения выручки по тарифам:

```python
revenue_per_month_smart.mean()
```

```python
revenue_per_month_ultra.mean()
```

Разница на первый взгляд существенная в пользу тарифа «Ультра», но мы работаем не с генеральными совокупностями. Требуется проверка гипотез:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(revenue_per_month_smart, revenue_per_month_ultra)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Действительно, гипотеза о равенстве выручки по тарифам не проходит. Принимаем альтернативную гипотезу "средняя выручка пользователей тарифов «Ультра» и «Смарт» различаются".
Для проверки новой альтернативной гипотезы "средняя выручка пользователей тарифов «Ультра» больше, чем у пользователей тарифов «Смарт» необходимо полученный p-value разделить пополам (т.е. выполнить односторонний тест):

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(revenue_per_month_smart, revenue_per_month_ultra)

print('p-значение:', results.pvalue / 2)

if results.pvalue / 2 < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Действительно, средняя выручка пользователей тарифов «Ультра» больше, чем у пользователей тарифов «Смарт».

### Проверка гипотезы об отличии средней выручки в Москве и в регионах

Необходимо проверить, действительно ли средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

**Нулевая гипотеза H₀**  
средняя выручка пользователей в Москве и в регионах одинаковая.  
**Альтернативная гипотеза H₁**  
средняя выручка пользователей в Москве и в регионах различается.

Сгруппируем пользователей по тарифам, создав два массива `revenue_per_month_msk` (выручка в Москве) и `revenue_per_month_regions` (выручка в регионах):

```python
revenue_per_month_msk = general_pivot[general_pivot['city'] == 'Москва']['revenue_per_month']
```

```python
revenue_per_month_regions = general_pivot[general_pivot['city'] != 'Москва']['revenue_per_month']
```

Рассмотрим средние значения выручки в Москве и в регионах:

```python
revenue_per_month_msk.mean()
```

```python
revenue_per_month_regions.mean()
```

Средняя выручка незначительно отличается, но мы работаем с выборками, а не генеральными совокупностями. Требуется проверка гипотез:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(revenue_per_month_msk, revenue_per_month_regions)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

p-value значительно выше порогового значения. Таким образом нельзя отклонять нулевую гипотезу "средняя выручка пользователей в Москве и в регионах одинаковая".
