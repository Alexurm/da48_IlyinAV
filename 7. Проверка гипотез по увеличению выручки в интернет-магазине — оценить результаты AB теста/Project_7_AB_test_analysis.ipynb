## Приоритизация гипотез

### Загрузка данных

```python
# импорт необходимых библиотек
import pandas as pd
import scipy.stats as stats
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
```

Полученная таблица содержит 9 гипотез по увеличению выручки с указанными параметрами:
* `Reach` (сколько пользователей затронет)
* `Impact` (насколько сильно повлияет изменение на пользователей)
* `Confidence` (насколько уверены, что изменение повлияет на пользователей именно так)
* `Effort` (сколько стоит протестировать гипотезу)

```python
# названия столбцов приведем в соответствии со змеиным регистром

hypothesis.columns = hypothesis.columns.str.lower()

#hypothesis.columns = ['hypothesis', 'reach', 'impact', 'confidence', 'efforts']
```

### Применение фреймворка ICE

Применим фреймворк `ICE` для приотеризации гипотез:

```python
# ICE = impact * confidence / efforts
hypothesis['ICE'] = round(hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts'], 2)
```

Отсортируем гипотезы по убыванию приоритета:

```python
hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False)
```

ТОП-3 приоритных гипотез по ICE:
* Запустить акцию, дающую скидку на товар в день рождения
* Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей
* Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок

### Применение фреймворка RICE

Применим фреймворк `RICE` для приотеризации гипотез:

```python
# RICE = reach * impact * confidence / efforts
hypothesis['RICE'] = hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence'] / hypothesis['efforts']
```

Отсортируем гипотезы по убыванию приоритета:

```python
hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False)
```

Гипотеза о запуске акции, дающей скидку на товар в день рождения, исчезла из приоритетных. ТОП-3 приоритных гипотез по RICE:
* Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок
* Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа
* Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей

### Заключение

Выполнена приоритизация гипотез по фреймворкам `ICE` и `RICE`, ТОП-3 гипотез представлены в сводной таблице:

Различия в списке приоритетных гипотез вызваны параметром `Reach`, используемым для расчета `RICE`. Рассмотрим параметр `Reach` перечисленных гипотез в порядке убывания:

За счет низкой оценки охвата пользователей (reach=1) гипотеза «Запустить акцию, дающую скидку на товар в день рождения» исчезла из ТОП-3 по `RICE`, а гипотеза «Добавить блоки рекомендаций товаров на сайт интернет магазина, чтобы повысить конверсию и средний чек заказа» за счет высокой оценки охвата пользователей (reach=8) вошла в ТОП-3 по `RICE`.

## Анализ A/B-теста

### Загрузка данных

```python
# загрузка данных
orders, visitors = (
    pd.read_csv('/datasets/orders.csv'), # данные о заказах
    pd.read_csv('/datasets/visitors.csv') # данные о пользователях
)
```

Рассмотрим последовательно загруженные данные и подготовим их для последующего анализа

В таблице отсутствуют пропущенные значения и явные дубликаты, однако необходимо выполнить следующее:
1. изменить названия стобцов в соответствии со змеиным регистром
2. преобразовать данные о времени

```python
# переименование столбцов
orders = orders.rename(columns={'transactionId': 'transaction_id',
                               'visitorId': 'visitor_id'})
```

```python
# преобразуем данные о времени
orders['date'] = pd.to_datetime(orders['date'])
```

```python
orders.info()
```

Названия колонок изменены, данные о времени приведены к необходимому типу.

Рассчитаем соотношение уникальных пользователей по группам A и B:

```python
orders[orders['group'] == 'A']['visitor_id'].nunique() / orders[orders['group'] == 'B']['visitor_id'].nunique()
```

Пользователей в группе А меньше на 14%, чем в группе B, что в итоге может повлиять на итоги исследования. Тем не менее, анализ A/B-теста будет выполняться по представленным данным.

```python
# создадим список с visitor_id для группы A
s_a = orders[orders['group'] == 'A']['visitor_id']

# создадим список с visitor_id для группы B
s_b = orders[orders['group'] == 'B']['visitor_id']

# объединим полученные visitor_id и удалим дубликаты
s_cross = pd.merge(s_a, s_b, how='inner', on=['visitor_id']).drop_duplicates()
```

Рассмотрим таблицу `orders` с уникальными пользователями в обеих группах:

```python
orders.query('visitor_id in @s_cross["visitor_id"]').sort_values(by='visitor_id')
```

### Стабильность кумулятивных метрик

#### Расчет кумулятивных данных

Создадим датафрейм `cumulative_data` с кумулятивными значениями по дням количества заказов, покупателей, выручки и посетителей интернет-магазина:

```python
# создаем массив уникальных пар значений дат и групп теста
dates_groups = orders[['date','group']].drop_duplicates() 

# получаем агрегированные кумулятивные по дням данные о заказах 
orders_аggregated = (dates_groups
                    .apply(lambda x: orders[np.logical_and(orders['date'] <= x['date'], orders['group'] == x['group'])]
                    .agg({'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', 'visitor_id' : 'nunique', 'revenue' : 'sum'}), axis=1)
                    .sort_values(by=['date','group'])
                   )

# получаем агрегированные кумулятивные по дням данные о посетителях интернет-магазина 
visitors_аggregated = (dates_groups
                       .apply(lambda x: visitors[np.logical_and(visitors['date'] <= x['date'], visitors['group'] == x['group'])]
                              .agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1)
                       .sort_values(by=['date','group'])
                      )

# объединяем кумулятивные данные в одной таблице и присваиваем ее столбцам понятные названия
cumulative_data = orders_аggregated.merge(visitors_аggregated, left_on=['date', 'group'], right_on=['date', 'group'])

# переименование столбцов
cumulative_data = cumulative_data.rename(columns={'transaction_id': 'orders',
                                                  'visitor_id': 'buyers'})
```

Кумулятивная выручка группы B в конце теста выше, чем у группы А. Тем не менее, необходимо будет проанализировать данные на предмет выбросов.

#### График кумулятивной конверсии по группам

### Расчет статистической значимости показателей по «сырым» данным

В данном разделе выполняется расчет статистической значимости различий в конверсии и в среднем чеке заказа между группами по «сырым» данным.
Необходимо проверить следующие гипотезы:
* конверсия группы B и группы A различается
* средний чек заказа группы B и группы A различается

Для первой гипотезы сформулированы нулевая и альтернативная гипотезы:  
**Нулевая гипотеза H₀**  
конверсия пользователей обеих групп одинаковая.  
**Альтернативная гипотеза H₁**  
конверсия пользователей обеих групп различается. 

Для второй гипотезы сформулированы нулевая и альтернативная гипотезы:  
**Нулевая гипотеза H₀**  
средний чек заказа обеих групп одинаковый.  
**Альтернативная гипотеза H₁**  
средний чек заказа обеих групп различается. 

Проверка будет выполняться непараметрическим тестом Манна-Уитни (метод `st.mannwhitneyu()`), для начала необходимо подготовить данные с накопленными значениями по группам.

#### Подготовка данных

Подготовим сводную таблицу для расчета всех основных показателей:

```python
# количество пользователей в день по группе A
visitors_a_daily = (visitors[visitors['group'] == 'A'][['date', 'visitors']]
                    .rename(columns={'visitors': 'visitors_per_date_a'})
                   )
# количество пользователей по группе A с накоплением
visitors_a_cummulative = (visitors_a_daily.apply(lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']]
                                               .agg({'date': 'max', 'visitors_per_date_a': 'sum'}), axis=1)
                          .rename(columns={'visitors_per_date_a': 'visitors_cummulative_a'})
                         )

# количество пользователей в день по группе B
visitors_b_daily = (visitors[visitors['group'] == 'B'][['date', 'visitors']]
                    .rename(columns={'visitors': 'visitors_per_date_b'})
                   )

# количество пользователей по группе B с накоплением
visitors_b_cummulative = (visitors_b_daily.apply(lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']]
                                               .agg({'date': 'max', 'visitors_per_date_b': 'sum'}), axis=1)
                          .rename(columns={'visitors_per_date_b': 'visitors_cummulative_b'})
                         )

# количество заказов и выручки в день для группы A 
orders_a_daily = (orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
                  .groupby('date', as_index=False)
                  .agg({'transaction_id': 'nunique', 'revenue': 'sum'})
                  .rename(columns={'transaction_id': 'orders_per_date_a', 'revenue': 'revenue_per_date_a'})
                 )

# количество заказов и выручки с накоплением для группы A
orders_a_cummulative = (orders_a_daily
                        .apply(lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']]
                               .agg({'date': 'max', 'orders_per_date_a': 'sum', 'revenue_per_date_a': 'sum'}), axis=1)
                        .sort_values(by='date')
                        .rename(columns={'orders_per_date_a': 'orders_cummulative_a', 
                                         'revenue_per_date_a': 'revenue_cummulative_a'})
                       )
                     
# количество заказов и выручки в день для группы B
orders_b_daily = (orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
                  .groupby('date', as_index=False)
                  .agg({'transaction_id': 'nunique', 'revenue': 'sum'})
                  .rename(columns={'transaction_id': 'orders_per_date_b', 'revenue': 'revenue_per_date_b'})
                 )

# количество заказов и выручки с накоплением для группы B
orders_b_cummulative = (orders_b_daily
                        .apply(lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']]
                               .agg({'date': 'max', 'orders_per_date_b': 'sum', 'revenue_per_date_b': 'sum'}), axis=1)
                        .sort_values(by='date')
                        .rename(columns={'orders_per_date_b': 'orders_cummulative_b', 
                                         'revenue_per_date_b': 'revenue_cummulative_b'})
                       )

# объединение данных
data = (orders_a_daily
        .merge(orders_b_daily, on='date', how='left') # присоединение количества заказов и выручки в день для группы B
        .merge(orders_a_cummulative, on='date', how='left') # присоединение количества заказов и выручки с накоплением для группы A
        .merge(orders_b_cummulative, on='date', how='left') # присоединение количества заказов и выручки с накоплением для группы B
        .merge(visitors_a_daily, on='date', how='left') # присоединение количества пользователей в день по группе A
        .merge(visitors_b_daily, on='date', how='left') # присоединение количества пользователей в день по группе B
        .merge(visitors_a_cummulative, on='date', how='left') # присоединение количества пользователей по группе A с накоплением
        .merge(visitors_b_cummulative, on='date', how='left') # присоединение количества пользователей по группе B с накоплением
       )
```

Названия столбцов данных датафрейма `data`:
* `date` — дата;
* `orders_per_date_a` — количество заказов в выбранную дату в группе A;
* `revenue_per_date_a` — суммарная выручка в выбранную дату в группе A;
* `orders_per_date_b` — количество заказов в выбранную дату в группе B;
* `revenue_per_date_b` — суммарная выручка в выбранную дату в группе B;
* `orders_cummulative_a` — суммарное число заказов до выбранной даты включительно в группе A;
* `revenue_cummulative_a` — суммарная выручка до выбранной даты включительно в группе A;
* `orders_cummulative_b` — суммарное количество заказов до выбранной даты включительно в группе B;
* `revenue_cummulative_b` — суммарная выручка до выбранной даты включительно в группе B;
* `visitors_per_date_a` — количество пользователей в выбранную дату в группе A;
* `visitors_per_date_b` — количество пользователей в выбранную дату в группе B;
* `visitors_cummulative_a` — количество пользователей до выбранной даты включительно в группе A;
* `visitors_cummulative_b` — количество пользователей до выбранной даты включительно в группе B.

#### Различие в конверсии между группами

Конверсия рассчитывается как отношение количества уникальных заказов к общему количеству пользователей. В `data` имеются столбцы по уникальному количеству заказов `orders_per_date_a` и `orders_per_date_b` для групп A и B соответственно, а также количество пользователей по группам в выбранную дату `visitors_per_date_a` и `visitors_per_date_b`.

Проведем двусторонний тест Манна-Уитни и найдем статистическую значимость в различии конверсии между группами:

```python
# конверсия по дням для группы A
conversion_per_date_a = data['orders_per_date_a'] / data['visitors_per_date_a']

# конверсия по дням для группы B
conversion_per_date_b = data['orders_per_date_b'] / data['visitors_per_date_b']

alpha = 0.05

results = stats.mannwhitneyu(conversion_per_date_a, conversion_per_date_b) #use_continuity=True, alternative='less')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
    
print('Разница в средней конверсии групп B и A составляет {0:.2%}'
      .format(conversion_per_date_b.mean()/conversion_per_date_a.mean()-1)) 
```

По «сырым» данным, несмотря на довольно значительную разницу в конверсии (17,5%) в пользу группы B, статистическая значимость по указанному параметру не достигается.

#### Различие в среднем чеке заказа между группами

Проведем двусторонний тест Манна-Уитни и найдем статистическую значимость в различии среднего чека между группами по «сырым» данным:

```python
alpha = 0.05

results = stats.mannwhitneyu(orders[orders['group']=='A']['revenue'], 
                             orders[orders['group']=='B']['revenue'])
                             #use_continuity=True, 
                             #alternative='less')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
    
print('Разница в среднем чеке групп B и A составляет {0:.2%}'
      .format(orders[orders['group']=='B']['revenue'].mean()/orders[orders['group']=='A']['revenue'].mean()-1))
```

По «сырым» данным, несмотря на довольно значительную разницу в среднем чеке (28,7%) в пользу группы B, статистическая значимость по указанному параметру не достигается.

### Расчет статистической значимости показателей по «очищенным» данным

В данном разделе выполняется расчет статистической значимости различий в конверсии и в среднем чеке заказа между группами по «очищенным» данным.

* средний чек заказа группы B и группы A различается

Для первой гипотезы сформулированы нулевая и альтернативная гипотезы:  
**Нулевая гипотеза H₀**  
конверсия пользователей обеих групп одинаковая.  
**Альтернативная гипотеза H₁**  
конверсия пользователей обеих групп различается. 

Для второй гипотезы сформулированы нулевая и альтернативная гипотезы:  
**Нулевая гипотеза H₀**  
средний чек заказа обеих групп одинаковый.  
**Альтернативная гипотеза H₁**  
средний чек заказа обеих групп различается. 

Проверка будет выполняться также непараметрическим тестом Манна-Уитни (метод `st.mannwhitneyu()`), для начала необходимо подготовить данные.

#### Подготовка данных

Ранее в разделе 2.3.3 был определен список аномальных пользователей `abnormal_users`, количество заказов или стоимость заказов которых вносят шум в данные, что может сказаться на корректности сравнения двух групп.

Исключим из датафрейма `orders` аномальных пользователей и сохраним в `orders_filtered`:

Подготовим сводную таблицу `data_filtered`:

```python
# количество заказов и выручки в день для группы A по очищенным данным
orders_a_daily_filtered = (orders_filtered[orders_filtered['group'] == 'A']
                           [['date', 'transaction_id', 'visitor_id', 'revenue']]
                           .groupby('date', as_index=False)
                           .agg({'transaction_id': 'nunique', 'revenue': 'sum'})
                           .rename(columns={'transaction_id': 'orders_per_date_a', 'revenue': 'revenue_per_date_a'})
                          )

# количество заказов и выручки в день для группы B по очищенным данным
orders_b_daily_filtered = (orders_filtered[orders_filtered['group'] == 'B']
                           [['date', 'transaction_id', 'visitor_id', 'revenue']]
                           .groupby('date', as_index=False)
                           .agg({'transaction_id': 'nunique', 'revenue': 'sum'})
                           .rename(columns={'transaction_id': 'orders_per_date_b', 'revenue': 'revenue_per_date_b'})
                          )

# объединение данных по группам
data_filtered = orders_a_daily_filtered.merge(orders_b_daily_filtered, on='date', how='left')

# объединение данных с количеством посещений из основного датафрейма по группам
data_filtered['visitors_per_date_a'] = data['visitors_per_date_a']
data_filtered['visitors_per_date_b'] = data['visitors_per_date_b']
```

Конверсия рассчитывается как отношение количества уникальных заказов к общему количеству пользователей.

В `data_filtered` имеются столбцы по уникальному количеству заказов `orders_per_date_a` и `orders_per_date_b` для групп A и B соответственно, а также имеется количество пользователей по группам в выбранную дату `visitors_per_date_a` и `visitors_per_date_b`.

Проведем двусторонний тест Манна-Уитни и найдем статистическую значимость в различии конверсии между группами:

```python
# конверсия по дням для группы A
conversion_per_date_a = data_filtered['orders_per_date_a'] / data_filtered['visitors_per_date_a']

# конверсия по дням для группы B
conversion_per_date_b = data_filtered['orders_per_date_b'] / data_filtered['visitors_per_date_b']

alpha = 0.05

results = stats.mannwhitneyu(conversion_per_date_a, conversion_per_date_b) #use_continuity=True, alternative='less')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
    
print('Разница в средней конверсии групп B и A составляет {0:.2%}'
      .format(conversion_per_date_b.mean()/conversion_per_date_a.mean()-1)) 
```

Достигнута статистическая значимость отличия конверсии между группами, средняя конверсия группы B выше. Проведем односторонний тест: 

```python
# конверсия по дням для группы A
conversion_per_date_a = data_filtered['orders_per_date_a'] / data_filtered['visitors_per_date_a']

# конверсия по дням для группы B
conversion_per_date_b = data_filtered['orders_per_date_b'] / data_filtered['visitors_per_date_b']

alpha = 0.05

results = stats.mannwhitneyu(conversion_per_date_a, conversion_per_date_b, use_continuity=True, alternative='less')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
    
print('Разница в средней конверсии групп B и A составляет {0:.2%}'
      .format(conversion_per_date_b.mean()/conversion_per_date_a.mean()-1)) 
```

По «очищенным» данным достигнута статистическая значимость, средняя конверсия группы B выше, чем группы A. Рост среднего значения конверсии составил в среднем 19,4%.

#### Различие в среднем чеке заказа между группами

Проведем двусторонний тест Манна-Уитни и найдем статистическую значимость в различии среднего чека между группами по «очищенным» данным:

```python
alpha = 0.05

results = stats.mannwhitneyu(orders_filtered[orders_filtered['group']=='A']['revenue'], 
                             orders_filtered[orders_filtered['group']=='B']['revenue']) 
                             #use_continuity=True, 
                             #alternative='less')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу: разница статистически значима')
else:
    print('Не получилось отвергнуть нулевую гипотезу, вывод о различии сделать нельзя') 
    
print('Разница в среднем чеке групп B и A составляет {0:.2%}'
      .format(orders_filtered[orders_filtered['group']=='B']['revenue'].mean()
              /orders_filtered[orders_filtered['group']=='A']['revenue'].mean()-1))
```

По «очищенным» данным статистическая значимость по разнице в среднем чека заказа не достигается, при этом средний чек группы B получился меньше на 4,85%, чем у группы A.
