Определим границы нормального размаха по годам для последующего анализа:

```python
#минимальное и максимальное значение года выпуска:
year_of_release_min = df['year_of_release'].min()
year_of_release_max = df['year_of_release'].max()

# первый квартиль:
q1 = df['year_of_release'].quantile(0.25)

# третий квартиль:
q3 = df['year_of_release'].quantile(0.75)

# межквартильный размах:
iqr = q3 - q1

# нижняя граница нормальных значений:
low_year_of_release = max(year_of_release_min, q1 - 1.5 * iqr)

# верхняя граница нормальных значений:
top_year_of_release = min(year_of_release_max, q3 + 1.5 * iqr)

print(f'Нормальные значения года выпуска игр находятся в диапазоне ({low_year_of_release}:{top_year_of_release})')
```

Выполним фильтрацию датафрейма `df_actual` по условию:

Уже из доступных данных можно сделать предварительный вывод о том, что в среднем платформа является актуальной порядка 7-13 лет.

Построим графики распределения глобальных продаж по этим платформам, в качестве интерполяции будем использовать полином 2-й степени:

```python
fig, ax = plt.subplots(3, 2, figsize=(18, 14), constrained_layout=True)
sns.set_style('whitegrid')

# цикл для перебора данных по платформам:
for platform, ax in zip(platforms_top.index, ax.flatten()):
    ax.set_title(platform)
    # определение области построения графика:
    xlim_left = platform_pivot[platform_pivot['platform']==platform]['year_of_release'].min()
    xlim_right = xlim_left + 12
    ax.set_xlim(xlim_left, xlim_right)
    ax.set_xticks(np.arange(xlim_left, xlim_right + 0.1, step=1))
    ax.set_ylim(0, 250)
    ax.grid(b=True, which='both')
    # построение зависимости глобальных продаж от года для данной платформы:
    sns.regplot(x='year_of_release', 
                y='total_sales', 
                order=2, 
                scatter_kws={'s': 20},
                data=platform_pivot[platform_pivot['platform']==platform],
                ax=ax)
```

Полученные распределения указывают на то, что в среднем жизненный срок платформы составляет порядка 10 лет, а затем на смену приходит новое поколение. Например Sony Playstation исходя из этих данных обновляет свои платформы каждые 6 лет (PS, PS2, PS3). Определим численно значение жизненного срока платформы:

```python
platform_life = []

for platform in platforms_top.index:
    year_start = platform_pivot[platform_pivot['platform']==platform]['year_of_release'].min()
    year_end = platform_pivot[platform_pivot['platform']==platform]['year_of_release'].max()
    platform_life.append(year_end - year_start)

platform_life = pd.Series(platform_life)

print(f'Средний жизненный срок платформы составляет {int(platform_life.mean())} лет')
```

При этом из распределения также видно, что для каждой платформы есть период увеличения продаж, пик продаж и спад. Ориентировочно первые 4-5 лет наблюдается устойчивый сильный рост продаж, далее пик и не менее крутой спад.

Построим диаграммы размаха для самых популярных жанров, определим следущий сводный список:

```python
top_genres = ["Action", "Shooter", "Sports", "Role-Playing", "Misc", "Platform", "Racing", "Fighting"]
```

```python
fig, ax = plt.subplots(4, 2, figsize=(18, 24), constrained_layout=True)
sns.set_style('whitegrid')
platform_prospective = ["PC", "PS4", "XOne"]

# цикл для перебора данных по жанрам:
for genre, ax in zip(top_genres, ax.flatten()):
    ax.set_title(genre)
    # определение верхней границы:  
    y_top = []
    for platform in platform_prospective:
        # первый квартиль:
        q1 = df_actual.query('genre == @genre and platform == @platform')['total_sales'].quantile(0.25)
        # третий квартиль:
        q3 = df_actual.query('genre == @genre and platform == @platform')['total_sales'].quantile(0.75)
        # межквартильный размах:
        iqr = q3 - q1
        y_top.append(q3 + 1.5 * iqr)

    y_top = int(max(y_top)+1)
    
    ax.set_ylim([0, y_top])
    ax.grid(b=True, which='both')
    sns.boxplot(x='platform',
                y='total_sales',
                data = df_actual.query('genre == @genre and platform in @platform_prospective'),
                width=0.4,
                ax=ax)
```

Согласно медианным значениям, полученным на диаграммах размаха:
1. Жанры Action, Shooter и Racing показывают примерно одинаковые продажи на обоих платформах PS4 и XOne;
2. Жанры Sports, Platform показывают больше продаж на PS4, чем на XOne;
3. Жанры Role-Playing, Misc, Fihting показывают больше продаж на XOne, чем на PS4.

PC во всех случаях показывают меньше продаж.

## Проверка гипотез

Необходимо проверить следующие гипотезы:
* cредние пользовательские рейтинги платформ Xbox One и PC одинаковые;
* cредние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

*Немного теории:*  

Проверка гипотез начинается с формулировки нулевой гипотезы H₀.  
H₀ всегда формулируется так, чтобы использовать знак равенства. Исходя из H₀ формулируется альтернативная гипотеза H₁. Это утверждение, которое принимается верным, если отбрасывается H₀. Иными словами, H₁ для H₀ — это альтернативная гипотеза, которая противоречит нулевой гипотезе.

При сравнении двух выборок проверяемая нулевая гипотеза состоит в том, что обе эти выборки происходят из нормального распределенных генеральных совокупностей с одинаковыми средними значениями:  
*H₀: μ₁=μ₁*

В общем виде проверка этой гипотезы выполняется при помощи p-value (уровень значимости статистики разности), который оценивает вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна. p-value - это вероятность, которая приводит к результатам, таким же предельным, какими могут оказаться наблюдаемые результаты с учетом модели нулевой гипотезы.

p-value сравнивается с выбранным уровнем значимости α (альфа), общепринятые пороговые значения - 5% и 1%, я буду принимать **5%**. Выбор любого из двух является произвольным решением - в данной процедуре нет ничего, что будет гарантировать правильные решения в x% случаях.
Значение α (уровень значимости) - это порог "необычности" в случайной модели нулевой гипотезы, по которому мы либо не будем отклонять нулевую гипотезу, либо наоборот отклоним и выберем альтернативную гипотезу.

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам:  
`scipy.stats.ttest_ind (array1, array2, equal_var)`  
Импортируем библиотеку:

```python
# импорт библиотеки scipy.stats:
from scipy import stats as st
```

В среднем пользователи платформы Xbox One ставят оценку 6,5 (медианное значение - 6,8). Минимальная оценка - 1,6, максимальная - 9,2.

```python
plt.figure(figsize=(12, 6))
plt.title('Распределение рейтинга игр для Xbox One')
plt.xlabel('Оценка пользователей')
plt.ylabel('Количество оценок')
plt.xlim(0, 10)
plt.ylim(0, 35)
plt.xticks(np.arange(0, 10.1, step=1))

x = user_score_xone
plt.hist(x, bins=20)
```

Имеем датасет с отрицательной скошенностью (скошенный влево). Это также можно заметить по тому факту, что среднее значение пользовательского рейтинга меньше медианного.

Определим среднее количество, дисперсию и стандартное отклонение для пользовательского рейтинга.

*Необходимо принимать во внимание, что в связи с удалением пропущенных значений пользовательского рейтинга, исследуется не вся доступная (генеральная) совокупность, а только выборка*.

```python
# средний пользовательский рейтинг:
user_score_xone_mean = user_score_xone.mean()
user_score_xone_mean
```

```python
# дисперсия:
user_score_xone_variance_estimate = np.var(user_score_xone, ddof=1)
user_score_xone_variance_estimate
```

```python
# стандартное отклонение:
user_score_xone_standard_deviation = np.sqrt(user_score_xone_variance_estimate)
user_score_xone_standard_deviation 
```

Проверим правило трех сигм, чтобы понять - входят ли значения `user_score_xone` в промежуток по трем стандартным отклонениям в обе стороны:

```python
# счетчик значений для подсчета попаданий в диапазон трех стандартных отклонений в обе стороны:
count = 0

# нижняя граница диапазона:
user_score_xone_bottom_line = user_score_xone_mean - 3 * user_score_xone_standard_deviation
# верхняя граниза диапазона:
user_score_xone_top_line = user_score_xone_mean + 3 * user_score_xone_standard_deviation

# цикл для перебора значений:
for col_name, data in user_score_xone.items():
    if data > user_score_xone_bottom_line and data < user_score_xone_top_line:
        count += 1

# правило трех сигм - 99% всех значений находятся в промежутке шести стандартных отклонений:
if count / len(user_score_xone) >= 0.99:
    user_score_xone_3_sigma = 'Выполняется'
    print ('Правило трех сигм соблюдается, изучаемая величина распределена нормально.')
else:
    user_score_xone_3_sigma = 'Не выполняется'
    print ('Правило трех сигм не соблюдается, изучаемая величина не распределена нормально.')
```

Это означает, что есть наличие выбросов вне интервала. Рассмотрим значения ниже границы нормального распределения:

```python
user_score_xone[user_score_xone < user_score_xone_bottom_line]
```

Таких значений всего 2 штуки.

**Нулевая гипотеза H₀**

**Альтернативная гипотеза H₁**  
Средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(user_score_xone, user_score_pc)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

```python
plt.figure(figsize=(12, 6))
plt.title('Распределение рейтинга игр для жанра Sports')
plt.xlabel('Оценка пользователей')
plt.ylabel('Количество оценок')
plt.xlim(0, 10)
plt.ylim(0, 50)
plt.xticks(np.arange(0, 10.1, step=1))

x = user_score_sports
plt.hist(x, bins=20)
```

Имеем датасет с отрицательной скошенностью (скошенный влево). Это также можно заметить по тому факту, что среднее значение пользовательского рейтинга меньше медианного.

Определим среднее количество, дисперсию и стандартное отклонение для пользовательского рейтинга.

*Необходимо принимать во внимание, что в связи с удалением пропущенных значений пользовательского рейтинга, исследуется не вся доступная (генеральная) совокупность, а только выборка*.

```python
# средний пользовательский рейтинг:
user_score_sports_mean = user_score_sports.mean()
user_score_sports_mean
```

```python
# дисперсия:
user_score_sports_variance_estimate = np.var(user_score_sports, ddof=1)
user_score_sports_variance_estimate
```

```python
# стандартное отклонение:
user_score_sports_standard_deviation = np.sqrt(user_score_sports_variance_estimate)
user_score_sports_standard_deviation 
```

Проверим правило трех сигм, чтобы понять - входят ли значения `user_score_sports` в промежуток по трем стандартным отклонениям в обе стороны:

```python
# счетчик значений для подсчета попаданий в диапазон трех стандартных отклонений в обе стороны:
count = 0

# нижняя граница диапазона:
user_score_sports_bottom_line = user_score_sports_mean - 3 * user_score_sports_standard_deviation
# верхняя граниза диапазона:
user_score_sports_top_line = user_score_sports_mean + 3 * user_score_sports_standard_deviation

# цикл для перебора значений:
for col_name, data in user_score_sports.items():
    if data > user_score_sports_bottom_line and data < user_score_sports_top_line:
        count += 1

# правило трех сигм - 99% всех значений находятся в промежутке шести стандартных отклонений:
if count / len(user_score_sports) >= 0.99:
    user_score_sports_3_sigma = 'Выполняется'
    print ('Правило трех сигм соблюдается, изучаемая величина распределена нормально.')
else:
    user_score_sports_3_sigma = 'Не выполняется'
    print ('Правило трех сигм не соблюдается, изучаемая величина не распределена нормально.')
```

#### Проверка гипотезы

**Нулевая гипотеза H₀**  
Средние пользовательские рейтинги жанров `Action` и `Sports` одинаковые.  
**Альтернативная гипотеза H₁**  
Средние пользовательские рейтинги жанров `Action` и `Sports` различаются.

Рассмотрим средние оценки пользователей по жанрам:

```python
user_score_action_mean
```

```python
user_score_sports_mean
```

Разница в оценках, на первый взгляд, довольно существенная. Мы работаем не с генеральными совокупностями, требуется проверка гипотез:

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(user_score_action, user_score_sports)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Таким образом, делаем вывод о том, что средние пользовательские рейтинги жанров Action и Sports различаются.
