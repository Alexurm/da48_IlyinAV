### Этаж квартиры

Необходимо дополнить датафрейм столбцом, который отображает информацию об этаже в объявлении:
* первый;
* последний;
* другой.

```python
# функция для добавления информации об этаже:
def floor_category_info(row):
    floor = row['floor']
    floors_total = row['floors_total']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'
```

Для полного датафрейма с увеличением площади растет стоимость квадратного метра (скорее всего это влияние элитного жилья в центре), для нормализованного датафрейма коэффициент получился отрицательным, но близким к нулю. Это означает, что влияние площади недвижимости на стоимость квадратного метра практически не оказывает влияния.

**Число комнат**

Согласно полученным данным, чем больше комнат, тем дороже недвижимость (это также очевидно).  
Получены следующие коэффициенты корреляции:
* k= 0.410 - полный датафрейм
* k= 0.418 - датафрейм с нормальными значениями

Число комнат оказывает среднее влияние на стоимость. Однако, при рассмотрении зависимости стоимости квадратного метра от площади недвижимости получены следующие коэффициенты корреляции:
* k = 0.050 - полный датафрейм
* k = -0.247 - датафрейм с нормальными значениями

Коэффициент корреляции между количеством комнат и стоимостью квадратного метра для полного датафрейма близкий к нулю, а для нормализованного датафрейма отрицательный (т.е. чем больше комнат, тем меньше стоимость квадратного метра).

**Расположение на этаже**

Получены следующие результаты:
1. предложения о недвижимости на первых этажах имеют наименьшую стоимость;
2. недвижимость на промежуточных этажах дороже чем на последних, если рассматривать медианные значения как по всему датафрейму, так и по нормальным данным.
3. влияние "элитной" недвижимости на стоимость предложений на последних этажах ощущается довольно сильно, если рассматривать средние значения по всему датафрейму.
